```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
PATH_DRIVE = '/content/drive/My Drive/Proyectos Nuevos/ML/regression_gasconsumption_keras_cnn/'
```

```
y_5 = np.load(PATH_DRIVE + "y_5.npy", allow_pickle=True )#time serie raw data
```

Using a 7-days frecuency for data seasonality, time serie decompose is done using an additive model. Three componets describes the time serie: trend, seasonal and residual.

```
decompfreq = 7#interval
decomposition = seasonal_decompose( pd.Series(y_5), freq=decompfreq, model='additive')
decomposition.plot()
```

```
fig1= pyplot.figure(figsize=(20,8))
plt.plot(y_5, color='y', label="real")
plt.plot(decomposition.trend.values, color='g', label="trend")
plt.plot(decomposition.seasonal.values, color='r', label="seasonal")
plt.plot(decomposition.resid.values, color='c', label="residual")
```

Seasonal Decomposition disadvantages:

*   data loss for first observations
*   model not fit so well rapid changes in data
*   decomposition model assume that data seasonality is repetead for choosed period
